```python
# Define the keywords to count
keywords = ['Online resources', 'Online bootcamps', 'Books', 'In-person bootcamps',
            'Conferences', 'Hackathons', 'Meetup.com events', 'Workshops',
            'School']


# Create a dictionary to store the counts for each keyword
keyword_counts = {keyword: 0 for keyword in keywords}

# Loop through the 'Helpful YouTube Channels' column
for cell in filtered_df_copy['Learning Methods']:
    # Split the cell values by commas
    cell_keywords = [value.strip() for value in cell.split(',')]
    
    # Check for the presence of each keyword
    for keyword in keywords:
        keyword_counts[keyword] += cell_keywords.count(keyword)

# Print the total counts for each keyword
for keyword, count in keyword_counts.items():
    print(f'{keyword}: {count}')
```
